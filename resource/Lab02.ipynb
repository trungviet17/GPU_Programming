```python
%%writefile lab02.cpp

#include <iostream>

#ifdef __NVCC__
#include <cuda_runtime.h>
#endif

int main() {
#ifdef __NVCC__

    int deviceCount;
    cudaGetDeviceCount(&deviceCount);
    std::cout << "Number of available GPUs: " << deviceCount << std::endl;

    cudaDeviceProp props;
    cudaGetDeviceProperties(&props, 0);

    std::cout << "Device Name: " << props.name << std::endl;
    std::cout << "Number of Processors: " << props.multiProcessorCount << std::endl;
    std::cout << "Maximum number of parallel blocks per multiprocessor: " << props.maxThreadsPerMultiProcessor << std::endl;
    std::cout << "Maximum number of threads per block: " << props.maxThreadsPerBlock << std::endl;
    std::cout << "Maximum number of threads per multiprocessor: " << props.maxThreadsPerMultiProcessor << std::endl;
    std::cout << "Global Memory Size: " << props.totalGlobalMem << " bytes" << std::endl;
    std::cout << "L1 Cache Size: " << props.sharedMemPerBlock << " bytes" << std::endl;
    std::cout << "L2 Cache Size: " << props.sharedMemPerBlock << " bytes" << std::endl;
    std::cout << "L3 Cache Size: " << props.sharedMemPerBlock << " bytes" << std::endl;

    int registerPerBlock;
    cudaDeviceGetAttribute(&registerPerBlock, cudaDevAttrMaxRegistersPerBlock, 0);
    std::cout << "Number of Registers per Block: " << registerPerBlock << std::endl;

    int registersPerMultiprocessor;
    cudaDeviceGetAttribute(&registersPerMultiprocessor, cudaDevAttrMaxRegistersPerMultiprocessor, 0);
    std::cout << "Number of Registers per Multiprocessor: " << registersPerMultiprocessor << std::endl;

    int peakClockRate;
    cudaDeviceGetAttribute(&peakClockRate, cudaDevAttrClockRate, 0);
    std::cout << "Peak Clock Rate: " << peakClockRate << " kHz" << std::endl;

    int peakMemoryClockRate;
    cudaDeviceGetAttribute(&peakMemoryClockRate, cudaDevAttrMemoryClockRate, 0);
    std::cout << "Peak Memory Clock Rate: " << peakMemoryClockRate << " kHz" << std::endl;
#else
    std::cerr << "Run this code with CUDA." << std::endl;
#endif
    return 0;
}
```

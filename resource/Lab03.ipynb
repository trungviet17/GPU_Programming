```python
%%writefile lab03.cpp

#include <iostream>
#include <chrono>

int main() {
    const int N = 100000000; // Number of iterations
    double sum = 0.0;

    // Start timing
    auto start = std::chrono::high_resolution_clock::now();

    // Perform a computationally intensive task
    for (int i = 0; i < N; ++i) {
        sum += 1.0 / (i + 1);
    }

    // Stop timing
    auto end = std::chrono::high_resolution_clock::now();
    std::chrono::duration<double> elapsed = end - start;

    // Output the result and the time taken
    std::cout << "Sum: " << sum << std::endl;
    std::cout << "Time taken: " << elapsed.count() << " seconds" << std::endl;

    return 0;
}

```

```python
import subprocess
import time
import numpy as np


flags = ['-O0', '-O1', '-O2', '-O3', '-Ofast']

result = {}

for flag in flags:
  executable = f"benchmark_{flag.replace('-', '')}"
  compile_command = f"g++ {flag} -o {executable} lab03.cpp"

  compile_result = subprocess.run(compile_command, shell=True, capture_output=True, text=True)

  times = []
  for i in range(10):
    start_time = time.time()
    run_command = f"./{executable}"
    run_result = subprocess.run(run_command, shell=True, capture_output=True, text=True)
    end_time = time.time()

    times.append(end_time - start_time)

  result[flag] = np.median(times)
```

```python
import matplotlib.pyplot as plt


plt.figure(figsize=(10, 6))
plt.plot(result.keys(), result.values(), marker='o')
plt.xlabel("Optimization Flags")
plt.ylabel("Execution Time (seconds)")
plt.title("Execution Time vs. Optimization Flags")
plt.grid(True)
plt.show()
```

```python
%%writefile hello.cu

#include <cuda_runtime.h>
#include <stdio.h>

__device__ void print_hello(int id) {
    printf("Hello world %d\n", id);
}

__global__ void kernel() {
  int id = threadIdx.x + 1;
  if (id <= 10) {
    print_hello(id);
  }
}


int main() {
    kernel<<<1,10>>>();
    cudaDeviceSynchronize();
    return 0;
}

```
